## Feature selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import scipy.io
import plotly.express as px

from umap.umap_ import UMAP
from sklearn.manifold import TSNE
```

```python
data = scipy.io.loadmat('soundData.mat')['sound_data']
data_label = scipy.io.loadmat('Stim288.mat')['labels'][0]
data_labels=[]
for i in data_label:
    data_labels.append(i[0])
    
data_norm = StandardScaler().fit_transform(data)
```

### Full dataset

```python
umap_2d = UMAP(n_components=2, init='random', random_state=0)
umap_3d = UMAP(n_components=3, init='random', random_state=0)

proj_2d = umap_2d.fit_transform(data_norm)
proj_3d = umap_3d.fit_transform(data_norm)

fig_2d = px.scatter(
    proj_2d, x=0, y=1,
    color=data_labels
)
fig_3d = px.scatter_3d(
    proj_3d, x=0, y=1, z=2,
    color=data_labels
)
fig_3d.update_traces(marker_size=5)

fig_2d.show()
fig_3d.show()
```

```python
X_embedded = TSNE(n_components=2).fit_transform(data_norm)
fig_2d = px.scatter(X_embedded, x=0, y=1, color=data_labels)

tsne = TSNE(n_components=3, random_state=0)
projections = tsne.fit_transform(data_norm, )

fig_3d = px.scatter_3d(
    projections, x=0, y=1, z=2,
    color=data_labels
)

fig_2d.update_traces(marker_size=5)
fig_3d.update_traces(marker_size=5)

fig_2d.show()
fig_3d.show()
```
